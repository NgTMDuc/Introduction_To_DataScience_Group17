```python
from bs4 import BeautifulSoup
import urllib.request
from lxml import etree
import requests
import json
from json import dumps
import pandas as pd
```

```python
url_prefix = "https://www.nguyenkim.com/dien-thoai-di-dong/page-"
no_page = 7
```

```python
url_list = []
count=0

for i in range(no_page):
    url = url_prefix + str(i+1)
    print(url)

    page = urllib.request.urlopen(url)
    soup = BeautifulSoup(page, 'html.parser')

    product_grid = soup.find('div', 'nk-product-cate-style-grid')
    for product_link in product_grid.find_all('a', 'product-render'):
        url_list.append(product_link['href'])
        print("\t" + str(product_link['href']))
        count = count + 1

print(count)
```

```python
nguyenkim_data = []

for url in url_list:
  try:
    page_request = urllib.request.urlopen(url)
    soup = BeautifulSoup(page_request, 'html.parser')
    table = soup.find('div', {'id': 'productSpecification_fullContent'})
    table_body = table.find('tbody')

    # loop over table to find number of rows with '' in first column
    # endrows = 0
    # for tr in table.findAll('tr'):
    #     if tr.findAll('td')[0].get_text() in (''):
    #         endrows += 1

    rows = len(table.find_all('tr'))
    # rows -= endrows + 1 # there is a pernicious final row that begins with 'Rk'

    header = []
    header.append('link')
    header.append('giá')
    for th in table_body.find_all('td', 'title'):
        key = th.get_text()[0:-1]
        header.append(key)

    print(header)

    body = []
    body.append(url)
    body.append(soup.find('span','nk-price-final').get_text())
    for td in table_body.find_all('td', 'value'):
        key = td.get_text()[1:-1]
        body.append(key)

    print(body)

    res = {header[i]: body[i] for i in range(len(header))}
    print(res)

    nguyenkim_data.append(res)
  except:
    continue
```

```python
print(nguyenkim_data)
```
